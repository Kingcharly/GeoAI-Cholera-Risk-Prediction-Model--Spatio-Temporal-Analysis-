```python
import geopandas as gpd
import pandas as pd
import os
```

```python
lgas = pd.read_csv('../data/alllgas.csv')
data2024 = pd.read_csv('../data/2024data.csv')
data2025 = pd.read_csv('../data/2025data.csv')
data2023 = pd.read_csv('../data/2023data.csv')
data2022 = pd.read_csv('../data/2022data.csv')
data2021 = pd.read_csv('../data/2021data.csv')
```

```python
lgas.head()
```

```python
data2022.head()
```

```python
data2021.head()
```

```python
def merge_data(year_data, lgas):
    merged_dataset = pd.merge(lgas, year_data, on = ['LGA', 'State'], how = 'left')
    merged_dataset['Case'] = merged_dataset['Case'].fillna(0)
    return merged_dataset
```

```python
dataset2025 = merge_data(data2025, lgas)
dataset2023 = merge_data(data2024, lgas)
dataset2022 = merge_data(data2023, lgas)
dataset2021 = merge_data(data2022, lgas)
dataset2024 = merge_data(data2021, lgas)
```

```python
dataset2025['Year'] = dataset2025['Year'].fillna(2025)
dataset2023['Year'] = dataset2023['Year'].fillna(2023)
dataset2022['Year'] = dataset2022['Year'].fillna(2022)
dataset2021['Year'] = dataset2021['Year'].fillna(2021)
dataset2024['Year'] = dataset2024['Year'].fillna(2024)
```

```python
target_data = pd.concat([dataset2021, dataset2022, dataset2023, dataset2024, dataset2025], ignore_index=True)
```

```python
target_data = target_data.sort_values(by=['Year', 'State'], ascending=[False, True]).reset_index(drop=True)
```

```python
target_data.shape
```

```python
target_data.to_csv('../data/merged_dataset.csv', index=False)
```

## Merging with Input Features

```python
features = pd.read_csv('../data/Features.csv')
```

```python
dataset = pd.merge(features, target_data, left_on=['lganame', 'statename'], right_on=['LGA', 'State'], how='inner')
```
